```sql
SELECT *
FROM workspace.data.bright_tv_viewership;


SELECT *
FROM workspace.data.bright_tv_username;



---Join tables query
SELECT
v.UserID0,
u.Name,
u.Gender,
u.Age,
u.Province,
v.Channel2,
v.RecordDate2,
v.`Duration 2`
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u
ON v.UserID0 = u.UserID;

--Active users per on each month over time
SELECT
DATE_TRUNC('month', v.RecordDate2) AS month,
COUNT(DISTINCT v.UserID0) AS active_users
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY 1
ORDER BY 1;

--Get usage by Gender
SELECT
u.Gender,
COUNT(*) AS total_sessions,
AVG(v.`Duration 2`) AS avg_watch_time
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY u.Gender;

--Get usage by age group
SELECT
CASE
WHEN u.Age < 18 THEN 'Under 18'
WHEN u.Age BETWEEN 18 AND 25 THEN '18-25'
WHEN u.Age BETWEEN 26 AND 35 THEN '26-35'
WHEN u.Age BETWEEN 36 AND 50 THEN '36-50'
ELSE '50+'
END AS age_group,
COUNT(*) AS sessions
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY age_group
ORDER BY sessions DESC;

--Get usage by Province
SELECT
u.Province,
COUNT(*) AS sessions
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY u.Province
ORDER BY sessions DESC;

--Get Top channels, views and Avarage watch time
SELECT
v.Channel2,
COUNT(*) AS views,
AVG(v.`Duration 2`) AS avg_watch_time
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY v.Channel2
ORDER BY views DESC
LIMIT 10;

--Get consumptions patterns
SELECT
HOUR(v.RecordDate2) AS hour,
COUNT(*) AS sessions
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY hour
ORDER BY hour;

--Get Weekday and weekend sessions
SELECT
CASE
WHEN DAYOFWEEK(v.RecordDate2) IN (1,7) THEN 'Weekend'
ELSE 'Weekday'
END AS day_type,
COUNT(*) AS sessions
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY day_type;

--Get low consumption days
SELECT
DATE(v.RecordDate2) AS date,
COUNT(*) AS sessions
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY date
ORDER BY sessions ASC
LIMIT 10;

-- Users session total watch time (high and low)
SELECT
v.UserID0,
u.Province,
COUNT(*) AS sessions,
SUM(v.`Duration 2`) AS total_watch_time
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY v.UserID0, u.Province
ORDER BY total_watch_time DESC;

--Get inactive users more than 14 days(churn analysis)
SELECT
v.UserID0,
u.Name,
u.Province,
MAX(v.RecordDate2) AS last_active
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY v.UserID0, u.Name, u.Province
HAVING DATEDIFF(CURRENT_DATE, MAX(v.RecordDate2)) > 14;

-- Get social media handle activity count
SELECT
u.`Social Media Handle`,
COUNT(v.UserID0) AS activity
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
WHERE u.`Social Media Handle` IS NOT NULL
GROUP BY u.`Social Media Handle`
ORDER BY activity DESC;

--Count total users
SELECT COUNT(*) AS TotalUsers
FROM workspace.data.bright_tv_viewership 

--Count users by gender
SELECT Gender,
COUNT(*) AS count
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY 
u.Gender

--Count users by province
SELECT Province,
COUNT(*) AS count
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u ON v.UserID0 = u.UserID
GROUP BY 
u.Province

--Check viewership per channel
SELECT v.Channel2,
COUNT(*) AS TotalViews
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u 
ON v.UserID0 = u.UserID
GROUP BY v.Channel2
ORDER BY TotalViews DESC

--Check users who never watched M-Net
SELECT u.UserID
  FROM workspace.data.bright_tv_username u
WHERE UserID NOT IN (
  SELECT v.UserID0 UserID
FROM workspace.data.bright_tv_viewership v
  WHERE Channel2 = 'M-Net')

--User info joined tables with their total views
SELECT 
  u.UserID, 
  u.Name, 
  u.Surname, 
  COUNT(v.UserID0) AS TotalViews
FROM workspace.data.bright_tv_username u
LEFT JOIN workspace.data.bright_tv_viewership v 
  ON u.UserID = v.UserID0
GROUP BY u.UserID, u.Name, u.Surname;

--Get users by channel and gender
SELECT 
  v.Channel2, 
  u.Gender, 
  COUNT(*) AS ViewCount
FROM workspace.data.bright_tv_viewership v
JOIN workspace.data.bright_tv_username u 
  ON v.UserID0 = u.UserID
GROUP BY v.Channel2, u.Gender;

--Get views by day
SELECT 
  v.RecordDate2, 
  COUNT(*) AS Views
FROM workspace.data.bright_tv_viewership v
GROUP BY v.RecordDate2
ORDER BY v.RecordDate2;

--Get peak day for each channel
SELECT 
  v.Channel2, 
  v.RecordDate2, 
  COUNT(*) AS Views
FROM workspace.data.bright_tv_viewership v
GROUP BY v.Channel2, v.RecordDate2
ORDER BY v.Channel2, Views DESC;

--Get Top users by activity
SELECT 
  u.UserID, 
  u.Name, 
  COUNT(v.UserID0) AS ViewsCount
FROM workspace.data.bright_tv_username u
JOIN workspace.data.bright_tv_viewership v 
  ON u.UserID = v.UserID0
GROUP BY u.UserID, u.Name
ORDER BY ViewsCount DESC
LIMIT 10;

--Get view trends by age group
SELECT
  CASE
    WHEN u.Age < 20 THEN '<20'
    WHEN u.Age BETWEEN 20 AND 29 THEN '20-29'
    WHEN u.Age BETWEEN 30 AND 39 THEN '30-39'
    ELSE '40+'
  END AS AgeGroup,
  COUNT(v.UserID0) AS Views
FROM workspace.data.bright_tv_username u
JOIN workspace.data.bright_tv_viewership v 
  ON u.UserID = v.UserID0
GROUP BY AgeGroup;



```
